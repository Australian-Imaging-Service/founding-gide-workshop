## Notebook 1 — FileFormats

> **File:** `fileformats.ipynb`  
> **Kernel:** Python

Scientific workflows deal with a wide variety of file formats — NIfTI, DICOM, Bruker ParaVision, and many more. The [`fileformats`](https://github.com/ArcanaFramework/fileformats) library provides a typed, extensible Python representation of these formats, making it easy to validate, mock, sample and convert between them.

This notebook covers:

1. **Using existing format types** — MIME-like string identifiers, the type hierarchy (`Nifti1 → BinaryFile → FileSet`), and generating test data with `mock()` and `sample()`
2. **Defining a simple binary format** — subclassing `BinaryFile` and adding basic validation
3. **Validated formats with magic numbers** — using byte-level signatures to verify file integrity at load time
4. **Multi-file formats** — representing formats that span a binary data file and a separate header (e.g. Analyze, Bruker 2dseq)
5. **Real-world example: Bruker ParaVision** — a worked example of a complex vendor format
6. **Extra hooks and converters** — implementing `load`/`save` for in-memory access and registering format converters so the framework can automatically transform between types

## Notebook 2 — Pydra

> **File:** `pydra.ipynb`  
> **Kernel:** Python

[Pydra](https://pydra.readthedocs.io) is the dataflow execution engine that powers AIS pipelines. It handles parallelisation, caching, and execution across environments (local, HPC, containers) so that pipeline authors can focus on the processing logic.

This notebook covers:

1. **Pre-built task packages** — using tasks from `pydra-tasks-fsl`, `pydra-tasks-mrtrix3` and similar packages to build pipelines without writing wrapper code
2. **Wrapping a command-line tool** — defining a `ShellCommandTask` to expose an arbitrary binary as a typed Pydra task with documented inputs and outputs
3. **Custom Python tasks** — writing a `FunctionTask` for in-process Python processing steps
4. **Composing workflows** — connecting tasks into a `Workflow`, passing outputs between nodes, and running the resulting graph

## Notebook 3 — AIS Pipelines

> **File:** `ais-pipelines.ipynb`  
> **Kernel:** Bash

This notebook brings the previous two together in the context of the Australian Imaging Service. You will build Docker-containerised pipelines from YAML specifications, deploy them to a local XNAT instance for testing, and submit them to the AIS community repository.

This notebook covers:

1. **Exploring the community repository** — understanding the structure of the `pipelines-community` repository and the YAML pipeline specification format
2. **Building and deploying your first pipeline** — using `pydra2app make` to build the example *Zip* pipeline and install it into a local XNAT instance
3. **Building the FSL BET pipeline** — building a real neuroimaging pipeline including a tool licence
4. **Designing a new pipeline from scratch** — using `pydra2app bootstrap` to auto-generate a specification for `mri_synthstrip`, then editing, building and testing it
5. **Submitting to the community repository** — creating a pull request on GitHub to trigger automated acceptance and deployment of your pipeline onto AIS
6. **FrameTree datasets and project-specific licences** — defining XNAT datasets and injecting per-project configuration such as a FreeSurfer licence
